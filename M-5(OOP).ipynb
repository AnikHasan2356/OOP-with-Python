```python
# Multiple Inheritance
# Parent Class
class Phone:
    category = "Electronics"

    def __init__(self, model, camera, battery, battery_percentage=100):
        self.model = model
        self.camera = camera
        self.battery = battery
        self.battery_percentage = battery_percentage

    def charge(self, hour):
        print(f"Charge completed {hour}%")

    def capture(self, photo):
        if self.battery_percentage <= 0:
            print("Please plug in")
        else:
            self.battery_percentage -= photo
            print(f"Photo capture in {self.model}")


# Child Class 1
class Smartphone(Phone):
    def __init__(self, model, camera, battery, processor):
        super().__init__(model, camera, battery)
        self.processor = processor

    def charge(self, hour):
        super().charge(hour)
        print("Fast charging is going on")


# Child Class 2
class Cooling:
    def __init__(self, cooling):
        self.cooling = cooling

    def cooling_on(self):
        print(f"Cooling mode is on by {self.cooling}")


# Multiple Inheritance
class Smartphone_cooling(Smartphone, Cooling):
    def __init__(self, model, battery, camera,processor, cooling):
        Smartphone.__init__(self, model,camera, battery, processor)
        Cooling.__init__(self, cooling)


# Object
pro = Smartphone_cooling("Pro Max", 4000,40, "Snapdragon", "Nitrogen")

pro.charge(20)
pro.cooling_on()
pro.capture(5)
```

```python
#Polymorphism= many form/use one element in different way by overrriding

class Camera:
    def __init__(self,name):
        self.name=name
    def capture(self):
        print("A photo is captured")


class Smart(Camera):
    def __init__(self,name,resolution):
      super().__init__(name)
      self.resolution=resolution
#Overriding
    def capture(self):
        print("A photo is captured by a Phone")


class DSLR(Camera):
    def __init__(self,name,resolution):
      super().__init__(name)
      self.resolution=resolution
#Overriding
    def capture(self):
        print("A photo is captured by a DSLR")

class Drone(Camera):
    def __init__(self,name,resolution):
      super().__init__(name)
      self.resolution=resolution
#Overriding
    def capture(self):
        print("A photo is captured by a Drone")

phone=Smart("Apple","60m")
d=DSLR("A","3k")
dr=Drone("B","4k")

phone.capture()
d.capture()
dr.capture()
```
